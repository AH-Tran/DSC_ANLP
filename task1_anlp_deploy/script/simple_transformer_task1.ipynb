```python
#data_df = pd.read_csv("data/full_train_short.tsv",sep='\t')
data_df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/full_train_short.tsv",sep='\t')

le = LabelEncoder()
data_df['label'] = le.fit_transform(data_df.label.values)
le_name_mapping = dict(zip(le.classes_, le.transform(le.classes_)))

train_df = data_df[:3200]
eval_df = data_df[3200:3600]
test_df = data_df[3600:]
```
